```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./梯度压缩对回归性能影响.csv")
```

```python
plt.figure(figsize=(7, 6))
sns.set_palette("Paired")

sns.scatterplot(x="gradient compression rate(%)", y="label1", data=df)
sns.scatterplot(x="gradient compression rate(%)", y="label2", data=df)
sns.scatterplot(x="gradient compression rate(%)", y="label3", data=df)
sns.scatterplot(x="gradient compression rate(%)", y="label4", data=df)
sns.scatterplot(x="gradient compression rate(%)", y="label5", data=df)
sns.scatterplot(x="gradient compression rate(%)", y="label6", data=df)

sns.lineplot(x="gradient compression rate(%)", y="label1", data=df)
sns.lineplot(x="gradient compression rate(%)", y="label2", data=df)
sns.lineplot(x="gradient compression rate(%)", y="label3", data=df)
sns.lineplot(x="gradient compression rate(%)", y="label4", data=df)
sns.lineplot(x="gradient compression rate(%)", y="label5", data=df)
sns.lineplot(x="gradient compression rate(%)", y="label6", data=df)

plt.rcParams['font.sans-serif'] = ['Times New Roman']
plt.ylabel("R2 Values", fontsize=12)
plt.xlabel("gradient compression rate(%)", fontsize=12)
plt.yticks([0.6, 0.7, 0.8, 0.9, 1.0])

plt.legend(labels=["label-1", "label-2", "label-3", "label-4", "label-5", "label-6"], loc="lower left", fontsize=10)
#plt.savefig("./梯度压缩对回归性能的影响.svg", dpi=600, bbox_inches='tight')
plt.savefig("./梯度压缩对回归性能的影响.pdf", dpi=600, bbox_inches='tight')
```
