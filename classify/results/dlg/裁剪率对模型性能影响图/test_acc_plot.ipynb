```python
import pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
baseline_acc = np.load('./baseline_test_acc.npy')
dgc_0002_acc = np.load('./dgc0.002_test_acc.npy')
dgc_0005_acc = np.load('./dgc0.005_test_acc.npy')
dgc_0008_acc = np.load('./dgc0.008_test_acc.npy')
dgc_0010_acc = np.load('./dgc0.01_test_acc.npy')

df = pd.DataFrame(index=range(300), columns=['baseline_acc', 'dgc_acc(compression rate=0.2%)', 'dgc_acc(compression rate=0.5%)',
                                             'dgc_acc(compression rate=0.8%)', 'dgc_acc(compression rate=1.0%)'])
df['baseline_acc'] = baseline_acc
df['dgc_acc(compression rate=0.2%)'] = dgc_0002_acc
df['dgc_acc(compression rate=0.5%)'] = dgc_0005_acc
df['dgc_acc(compression rate=0.8%)'] = dgc_0008_acc
df['dgc_acc(compression rate=1.0%)'] = dgc_0010_acc
```

```python
plt.figure(figsize=(7, 6))
plt.rcParams['font.sans-serif'] = ['Times New Roman']
sns.lineplot(data=df)
plt.xlabel('epoch', fontsize=12)
plt.ylabel('accuracy', fontsize=12)
plt.legend(loc='lower right', fontsize=10)

plt.savefig('./test_acc_classify.pdf', dpi=600, bbox_inches='tight')
plt.show(block=True)
```
